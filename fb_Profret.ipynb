```
# dataframes creation for both training and testing datasets 
avocado_df = pd.read_csv('/content/drive/My Drive/DATASETS/avocado.csv')
```

```
# Select specific region
av_reg=avocado_df[avocado_df['region']=='West']
```

```
av_reg = av_reg.sort_values('Date')
```

```
plt.plot(av_reg['Date'],av_reg['AveragePrice'])
```

```
av_reg=av_reg.rename(columns={'Date':'ds','AveragePrice':'y'})
```

```
m = Prophet()
m.fit(av_reg)
# Forcasting into the future
future = m.make_future_dataframe(periods=365)
forecast = m.predict(future)
```

```
figure3 = m.plot( forecast, xlabel='Date',ylabel='Price')
```

```
figure4 = m.plot_components(forecast)
```
